## What's about "shares" ? How to interpret the problem

```python
print(df['shares'].describe()) # The median is 1400, we will use this value in order to consider popular/unpopular articles
```

We choose to consider this problem as a classification problem :

- We tried linear regression in order to predict a number of share for any article, but the MSE were huge and the score was under 0.15. We won't show the code results since the notebook is already big.

- We will consider a post as to be popular if the number of shares > the median, else it will be considered as unpopular. The problem is a classification problem. We decided not to create more than 2 classes since the accuracy would drasticaly decrease then. 
- We have already tested with the thresh old set to 3355 which is the mean, the best accuracy we had was ~80.3%, but the recall and F1-score was around 0 with every model. Indeed, the model was able to recognize an article was unpopular very well (because it had many of this examples), but when it had to predict that a popular article was popular it sucked really much (because there was not many cases). 
- We decided not to provide you the manipulation around the 3355 thresh old since the Notebook would be twice as long. Though, we put an example at the end of the Notebook in order to prove our work.

```python
liste = list()
for x in df['shares']:
    if x >= 1400:
        liste.append(4)
    else:
        liste.append(2)

df["popularity"] = liste
```

- "4" : the article is popular, "2" : unpopular

- We are always going to compare and optimize models (by tweaking the hyperparameters) based on their accuracy score and print and compare their f1 score, roc auc score, precision score and recall score for the hyperparameters that optimized the best the accuracy. 

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, precision_score, recall_score
```

- We create a 80% of the data training set and 20% test set

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,:-2].values
Y = df.iloc[:, -1].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

Using methods not needing scaling on variables : 

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
values = [185, 190,195, 200, 250, 300, 350, 400, 500]
accuracy = []
for n in values:    
    clf = GradientBoostingClassifier(n_estimators=n, max_depth=5,
                                 random_state=0)
    clf.fit(X_train, Y_train)  
    Y_pred = clf.predict(X_test)
    accuracy.append(100*accuracy_score(Y_test, Y_pred))

plt.figure(figsize=(10,7))
plt.plot(values, accuracy, 'r-', label='Gradient Boosting Accuracy by number of estimator')
plt.plot(values, accuracy, 'bx')
plt.xlabel('Number of estimators')
plt.ylabel('Gradient Boosting Accuracy')
plt.legend()
plt.grid()
plt.title('Gradient Boosting Accuracy by number of estimator')
plt.show()

#We have tested with several values in the range of 1 to 500
```

```python
#Here we are zooming on the values which are close to each other
#I have manullay checked that 188, 189 etc were decreasing too

values = [190, 191, 192, 193, 195]
accuracy = []
for n in values:    
    clf = GradientBoostingClassifier(n_estimators=n, max_depth=5,
                                 random_state=0)
    clf.fit(X_train, Y_train)  
    Y_pred = clf.predict(X_test)
    accuracy.append(100*accuracy_score(Y_test, Y_pred))

plt.figure(figsize=(10,7))
plt.plot(values, accuracy, 'r-', label='Gradient Boosting Accuracy by number of estimator')
plt.plot(values, accuracy, 'bx')
plt.xlabel('Number of estimators')
plt.ylabel('Gradient Boosting Accuracy')
plt.legend()
plt.grid()
plt.title('Gradient Boosting Accuracy by number of estimator')
plt.show()

#We cleary see that the best accuracy comes for n_estimators = 192
```

## Ada Boost Classifier

Pros

AdaBoost is easy to implement. It iteratively corrects the mistakes of the weak classifier and improves accuracy by combining weak learners. You can use many base classifiers with AdaBoost. AdaBoost is not prone to overfitting. This can be found out via experiment results, but there is no concrete reason available.

Cons

AdaBoost is sensitive to noise data. It is highly affected by outliers because it tries to fit each point perfectly. AdaBoost is slower compared to XGBoost.
